```python
def convert_cast(casts):
    list = casts.split(", ")
    return list

df['cast'] = df['cast'].apply(convert_cast)

df.head()
```

```python
df['genre'] = df['genre'].apply(lambda x: [i.replace(" ","") for i in x])
df['cast'] = df['cast'].apply(lambda x: [i.replace(" ","") for i in x])
df['director'] = df['director'].apply(lambda x: [i.replace(" ","") for i in x])

df.head()
```

```python
df['genre'] = df['genre'].apply(lambda x: [i.lower() for i in x])
df['cast'] = df['cast'].apply(lambda x: [i.lower() for i in x])
df['director'] = df['director'].apply(lambda x: [i.lower() for i in x])
df['overview'] = df['overview'].apply(lambda x: [i.lower() for i in x])
df['year'] = df['year'].apply(lambda x: str(x)).apply(lambda x: x.split(" "))



df.head()
```

```python
df['tag'] = df['year'] + df['genre']+df['overview']+df['director']+df['cast'] 
df['tag'] = df['tag'].apply(lambda x: " ".join(x))

print(df['tag'])
```

```python
# ! pip install nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem_words(text):
    list =[]
    for ele in text.split(" "):
        list.append(ps.stem(ele))
    return " ".join(list)

df['tag'].apply(stem_words)
df['tag']
```

```python
! pip install scikit-learn

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(df['tag']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_vector = cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index = df[df['movie_name'] == movie].index[0]
    distances = similarity_vector[movie_index]
    moivie_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x:x[1])[1:6]

    for i in moivie_list:
        print(df.iloc[i[0]].movie_name)

recommend('Pathaan')
```
